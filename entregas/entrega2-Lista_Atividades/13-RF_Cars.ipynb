### 13. Regressão com Random Forest

Crie um modelo de Random Forest para prever o preço de carros.

Compare os resultados com um modelo de regressão linear usando MSE e R²

```python
# DATASET UTILIZADO -> CarsDetails (data/CarsDetails.csv)
```

```python
# Conjunto de Importações
import pandas as pd
import numpy as np
```

```python
dados = pd.read_csv("data/carsDetails.csv")            # leitura do .csv
dados.head()                                            # cabeçalho do .csv
```

TRATAMENTO DOS DADOS

```python
# Remoção de dados desnecessários:
dados = dados.drop(['Model'], axis=1)
dados = dados.drop(['Location'], axis=1)
dados = dados.drop(['Engine'], axis=1)
dados = dados.drop(['Max Power'], axis=1)
dados = dados.drop(['Max Torque'], axis=1)
```

```python
# MAPEAMENTOS DOS DADOS PARA NUMÉRICOS
dados['Seller Type'], cars_dataset_seller_mapping = pd.factorize(dados['Seller Type'])
dados['Owner'], cars_dataset_owner_mapping = pd.factorize(dados['Owner'])
dados['Color'], cars_dataset_color_mapping = pd.factorize(dados['Color'])
dados['Transmission'], cars_dataset_transmission_mapping = pd.factorize(dados['Transmission'])
dados['Fuel Type'], cars_dataset_fuel_mapping = pd.factorize(dados['Fuel Type'])
dados['Make'], cars_dataset_make_mapping = pd.factorize(dados['Make'])
dados['Drivetrain'], cars_dataset_drive_mapping = pd.factorize(dados['Drivetrain'])
dados = dados.dropna(subset=['Make', 'Year', 'Kilometer', 'Fuel Type', 'Color', 'Owner', 'Seller Type', 'Drivetrain', 'Length', 'Width', 'Height', 'Seating Capacity', 'Fuel Tank Capacity'])

dados.head()            # representação toda numérica
```

```python
# Separando em amostra de Treino e Teste
from sklearn.model_selection import train_test_split

X = dados.drop('Price', axis=1)          # "medv" é a coluna alvo
Y = dados['Price']                       # Coluna alvo

X_treino,X_teste,Y_treino,Y_teste=train_test_split(X,Y,test_size=0.30,random_state=0)           # separação de 30% para teste e 70% para treino
```

```python
# Aplicando Modelo de RF (Random Forest)
from sklearn.ensemble import RandomForestRegressor

modelo_RF = RandomForestRegressor()                     # uso do modelo DecisionTreeRegressor()

modelo_RF.fit(X_treino, Y_treino)                       # treinamento do modelo
```

```python
# Realizando a predição com os dados de validação:
y_pred = modelo_RF.predict(X_teste)
```

Regressão Linear (COMPARAÇÃO)


```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()

lr_model.fit(X_treino, Y_treino)

lr_pred = lr_model.predict(X_teste)
```

# RESULTADOS FINAIS (MSE e R²)

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# MSE
rf_mse = mean_squared_error(Y_teste, y_pred)
lr_mse = mean_squared_error(Y_teste, y_pred)
# R² SCORE
r2score = r2_score(y_pred,Y_teste)     
r2score_lr = r2_score(lr_pred, Y_teste)

print("RANDOM FOREST:")
print(f"MSE score: {rf_mse}")
print(f"R² score: {r2score}")
print("--------------------")
print("Regressão Linear:")
print(f"MSE Score: {lr_mse}")
print(f"R² Score: {r2score_lr}")
```
